```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
# %matplotlib notebook
```

####  1) Na base do SINASC do DataSUS, faça um boxplot da idade da mãe.

```python
df_sinasc = pd.read_csv('SINASC_RO_2019.csv')
df_sinasc[['IDADEMAE']].boxplot()
```

```python
sns.boxplot(y="IDADEMAE", data=df_sinasc[['IDADEMAE']])
```

####  2) Faça um boxplot do peso do bebê

```python
sns.boxplot(y="PESO", data=df_sinasc[['PESO']])
```

####  3) Faça um histograma da idade da mãe

```python
sns.displot(data = df_sinasc[['IDADEMAE']]
            ,alpha = .25
            ,x = 'IDADEMAE'
            ,bins = 20
            ,element = 'step'
            ,kde = True
           )
```

#### 4) Faça um histograma do peso do bebê

```python
sns.displot(data = df_sinasc[['PESO']]
            ,alpha = .25
            ,x = 'PESO'
            ,bins = 20
            ,element = 'step'
            ,kde = True
           )
```
