#### 1) A base 'sinasc_RO_2019.csv' contém dados do sistema de nascidos vivos do DataSUS, cujo dicionário de dados está no arquivo 'Estrutura_SINASC_para_CD.pdf'. Carregue esse arquivo.

```python
arquivo = 'sinasc_RO_2019.csv'

sin = pd.read_csv(arquivo)

sin.head(5)
```

#### 2) Categorize a variável APGAR5 conforme as seguintes quebras:

- 0 a 3: 'asfixia grave'
- 4 a 5: 'asfixia moderada'
- 6 a 7: 'asfixia leve'
- 8 a 10: 'normal'

```python
sin.loc[(sin['APGAR5'] <= 3), 'APGAR5_cat'] = 'asfixia grave'
sin.loc[(sin['APGAR5'] >= 4) & (sin['APGAR5'] <= 5), 'APGAR5_cat'] = 'asfixia moderada'
sin.loc[(sin['APGAR5'] >= 6) & (sin['APGAR5'] <= 7), 'APGAR5_cat'] = 'asfixia leve'
sin.loc[(sin['APGAR5'] >= 8), 'APGAR5_cat'] = 'normal'
```

```python
sin['APGAR5_cat'].unique()
```

#### 3) Monte um gráfico comparando a distribuição de APGAR5 por sexo para cada grupo de tempo de gestação (GESTACAO).

```python
tempodegestacao = [['Menos de 22 semanas', '22 a 27 semanas', '28 a 31 semanas'],
                 ['32 a 36 semanas', '37 a 41 semanas', '42 semanas e mais']]

plt.rc('figure', figsize = (20, 10))

fig, axes = plt.subplots(2, 3)

for i in range(len(tempodegestacao)):
    for j in range(len(tempodegestacao[i])):
        sns.countplot(ax = axes[i, j]
                      , x = 'APGAR5'
                      , hue = 'SEXO'
                      , data = sin[sin['GESTACAO'] == tempodegestacao[i][j]]
                     ).legend(loc = 'upper left').set(title=tempodegestacao[i][j])
```

####  4) Monte uma tabela cruzada com APGAR5 nas colunas e GESTACAO nas linhas

```python
tabcruz = pd.crosstab(sin['GESTACAO'], sin['APGAR5'])
tabcruz
```

####  5) Da tabela cruzada do intem anterior, calcule os percentuais de linha. Ou seja, a distribuição da variável APGAR5 para cada classe de GESTACAO dado.

```python
tabcruz_pctlin = tabcruz.div(tabcruz.sum(axis = 1), axis = 0)
tabcruz_pctlin
```

####  6) Os médicos costumam classificar parto prematuro quando a gestação tem menos que 37 semanas. Faça um gráfico de barras da tabela cruzada acima e conclua se você acha que este conjunto de dados corrobora ou não com a afirmação dos médicos.

```python
tabcruz_cat = pd.crosstab(sin['GESTACAO'], sin['APGAR5_cat'])
tabcruz_cat_pctlin = tabcruz_cat.div(tabcruz.sum(axis = 1), axis = 0)
tabcruz_cat_pctlin.plot.bar()
```

Pelo gráfico de barras nota-se que que o quantitativo de RN com grave começa a ser registrados em nascimentos de RN com idade gestacional abaixo de 37 semanas, nota-se que quando menor a idade gestacional a gravidade aumenta, classifcando a gestação como uma gestação de risco

#### 7) Veja se há relação entre APGAR5 e o peso do bebê. Isso pode ser feito de diversas formas.
```python
tempodegestacao = [['Menos de 22 semanas', '22 a 27 semanas'], 
                 ['28 a 31 semanas', '32 a 36 semanas'], 
                 ['37 a 41 semanas', '42 semanas e mais']]

plt.rc('figure', figsize = (20, 20))

fig, axes = plt.subplots(3, 2)

for i in range(len(tempodegestacao)):
    for j in range(len(tempodegestacao[i])):
        sns.histplot(ax = axes[i, j]
                     , x = 'PESO'
                     , hue = 'APGAR5_cat'
                     , hue_order = ['normal', 'asfixia leve', 'asfixia moderada', 'asfixia grave']
                     , element = 'step'
                     , data = sin[(sin['GESTACAO'] == tempodegestacao[i][j]) 
                                        & (sin['PESO'] < 3000)]
                    ).set_title(tempodegestacao[i][j])
```

No gráfico nota-se que as asfixa graves tem maior numero registrados em RN de tempo de gestão menor que 27 semnas e com peso abaixo de 1 000 gramas. Encontramos, também registro em RC de 32 a 36 semanas com peso meno que 100o gramasm, assim como os que são maior que 42 semans (Pós termo) porém ambos que menor frequeência.
