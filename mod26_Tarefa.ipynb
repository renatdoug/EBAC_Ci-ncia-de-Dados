# Tarefa - Álgebra Linear

A célula abaixo gera dados de notas de duas provas, p1 e p2, conforme em aula, mas com mais alunos.

## 1. Reduzindo correlações

a) Já vimos que expressar os dados através da média e diferença é uma transformação linear. Obtenha a média e a diferença através de uma multiplicação matricial, conforme visto em aula.

b) Calcule a correlação entre os dados da **p1** e **p2**

c) Calcule a correlação entre os dados transformados (média e diferença)

d) Comente as correlações observadas.

```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

# Gerando os dados
np.random.seed(123)

p1 = np.random.random(50) *10
p1[p1>10] = 10
p1[p1<0] = 0

p2 = p1 + np.random.normal(0,1,50)
p2[p2>10] = 10
p2[p2<0] = 0

df = pd.DataFrame({'p1': p1, 'p2': p2})
```

```python
#a) 
df['média'] = df.mean(axis=1)
df['evolução'] = df.p2-df.p1
df.style.format(precision=1, decimal=',')
```

```python
#b)
corr_matrix = np.corrcoef(df['p1'], df['p2'])
correlacao = corr_matrix[0, 1]

print(f"A correlação entre 'p1' e 'p2' é: {correlacao}")
```

```python
#C 
corr_matrix = np.corrcoef(df['média'], df['evolução'])
correlacao = corr_matrix[0, 1]

print(f"A correlação entre 'média' e ''evolução' é: {correlacao}")
```

#d

Temos uma correlação forte e positiva entre 'p1' e 'p2', indicando uma relação linear forte entre essas variáveis. Por outro lado, a correlação entre 'média' e 'evolução' é muito fraca, indicando uma relação linear praticamente inexistente entre essas variáveis.

## 2. Soma e diferença

Realize agora uma transformação semelhante, mas que ao invés da média, dê a soma da **p1** e **p2**. Construa o gráfico e conclua que diferenças visuais você observa.

```python
df['soma'] = df.p2+df.p1
df.style.format(precision=1, decimal=',')
```

```python
#Definindo figura e configurando eixos
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(1,1,1)

xmin = -5
xmax = 20
ymin = -5
ymax = 5

# ax.axis('equal')
ax.set_aspect('equal', 'box')
ax.set(xlim=(xmin, xmax), ylim=(ymin, ymax))

lista_xticks = list(range(int(xmin), int(xmax)+1))
ticksx = ax.set_xticks(lista_xticks)
labelsx = ax.set_xticklabels(lista_xticks, fontsize='large')

lista_yticks = list(range(int(ymin), int(ymax)+1))
ticksy = ax.set_yticks(lista_yticks)
labelsy = ax.set_yticklabels(lista_yticks, fontsize='large')

#linhas horizontais e verticais de referência
ax.axvline(0, linewidth=.5, linestyle='-')
ax.axhline(0, linewidth=.5, linestyle='-')

# Plotar a média
plt.scatter(df['média'], df['evolução'], label='média')

# Plotar a soma
plt.scatter(df['soma'], df['evolução'], label='soma')

#Legenda
plt.legend()
```

```python
corr_matrix = np.corrcoef(df['soma'], df['evolução'])
correlacao = corr_matrix[0, 1]

print(f"A correlação entre 'média' e ''evolução' é: {correlacao}")
```

   A concentração das médias em um intervalo menor sugere uma distribuição mais consistente das notas médias entre os alunos, enquanto a dispersão das somas indica uma variação maior nas notas individuais.
    A correlação próxima de zero entre a média e a diferença das notas sugere que a média das notas não está diretamente relacionada à diferença entre as notas das avaliações p1 e p2.
