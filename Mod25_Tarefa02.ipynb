1- Calcule de forma numrica a integral da funo $x^3$ avaliada entre os pontos $1$ e $2$. Identifique o nmero de retngulos que voc usou.  




2- Calcule essa rea de forma analtica.

```python
import pandas as pd
import numpy as np

f = lambda x: x**3

def calcula_area(a, b, func, qtd_ret):
    delta = (b-a)/qtd_ret # largura do retngulo
    x_cols = np.linspace(a, b-delta, qtd_ret)+delta/2 # Diviso do eixo X
    y_cols = func(x_cols) # alturas dos retngulos
    return sum(delta*y_cols)
```

```python
#Clculos para diferentes nmeros de retngulos;

qtd_ret = [4,8,12,16,20,24, 28, 32, 36, 40, 44,48,50,100]
for n in qtd_ret:
    print(f"Resultado considrando {n} quadrados  {calcula_area(1,2,f,n):.2f}")
```
