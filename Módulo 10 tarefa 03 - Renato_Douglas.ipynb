````python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
````

````python
df_draw = pd.read_csv('previsao_de_renda.csv')

df = df_draw.copy()

df.drop(['Unnamed: 0', 'index'], axis = 1, inplace = True)

df.head()
````

````python
df['data_ref'] = pd.to_datetime(df['data_ref'])

df.info()
````

1) Construa um gráfico de barras para avaliar a distribuição das variáveis qualitativas no tempo. Você acha que houve alguma alteração importante a ser investigada?

````python
# Lista de variáveis qualitativas
variables = ['sexo', 'posse_de_veiculo']

for var in variables:
    _, ax = plt.subplots(figsize=(20, 10))
    
    ax = sns.countplot(x='data_ref', hue=var, data=df)
    
    tick_labs = df['data_ref'].map(lambda ts: ts.strftime("%m-%Y")).unique()
    
    ticks = ax.set_xticks(list(range(df['data_ref'].nunique())))
    labels = ax.set_xticklabels(tick_labs, rotation=0)  # A rotação foi ajustada para 45 graus para melhor visualização, caso as labels sejam muitas e possam se sobrepor.
    
    plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0)
    plt.title(f"Distribuição de {var} ao longo do tempo")
    plt.show()

````

2) Para cada variável qualitativa, construa um gráfico mostrando as médias de cada nível da variável no tempo e avalie a estabilidade. Houve alguma alteração importante? Você acha que se pode utilizar essas variáveis em uma modelagem?

````python

variaveis_qualitativas = ['sexo', 'posse_de_veiculo', 'posse_de_imovel', 'tipo_renda', 
                          'educacao', 'estado_civil', 'tipo_residencia', 'mau']

n_vars = len(variaveis_qualitativas)
n_cols = 2
n_rows = (n_vars // n_cols) + (n_vars % n_cols > 0)

fig, axes = plt.subplots(n_rows, n_cols, figsize=(20, 5*n_rows))
axes = axes.ravel() 

for i, variavel in enumerate(variaveis_qualitativas):
    ax = sns.pointplot(x="data_ref", y="renda", hue=variavel, data=df, dodge=True, errorbar=('ci', 95), ax=axes[i])

    tick_labs = df['data_ref'].map(lambda ts: ts.strftime("%m-%Y")).unique()
    ticks = ax.set_xticks(list(range(df['data_ref'].nunique())))
    labels = ax.set_xticklabels(tick_labs, rotation=45)  
    
    ax.legend(loc = 'lower center', bbox_to_anchor=(0.5, -.50),ncol=3)
    ax.set_title(f'Média da Renda por {variavel} ao Longo do Tempo')

# Esconde os eixos adicionais (se houver)
for j in range(i+1, n_cols*n_rows):
    axes[j].axis('off')

plt.tight_layout()
plt.show()
````

A plotagagem das variáveis num gráfico de erro padrão renda versus tempo, mostra que que mudança continua ainda com a variável educação, porém com outra subcategoria, que é a pós graduação, mostrando que pessoas mais especializadas podem apresentar uma renda um pouco acima da média geral. 

Considerando que as alterações observadas são pontuais, poderíamos, sim essas variáveis em uma modelagem
