1) Com o que você viu em aula, faça uma análise das variáveis qualitativas (posse de imóvel e posse de veículo) para verificar se são boas preditoras de *renda*. Alguma delas parece discriminar *renda* de forma estatisticamente significante?

````python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
````

````python
df_raw = pd.read_csv('previsao_de_renda.csv')

df_raw.head()
````

````python
df = df_raw.copy()

df.drop(['Unnamed: 0', 'mau', 'index'], axis = 1, inplace = True)

df.head()
````

````python
plt.figure(figsize=(10, 6))

# Posse de Imóvel
ax = sns.pointplot(x='posse_de_imovel', y='renda', 
                   data=df, dodge=True, errorbar=('ci', 90),
                   color='blue', label='Posse de Imóvel')

# Posse de Veículo
ax = sns.pointplot(x='posse_de_veiculo', y='renda', 
                   data=df, dodge=True, errorbar=('ci', 90),
                   color='orange', label='Posse de Veículo')

ax.set_title('Relação entre Posse de Imóvel, Posse de Veículo e Renda')
ax.set_xlabel('Posse')
ax.set_ylabel('Renda')
ax.legend()

plt.show()
````

Apártir dos gráficos gerados, é observável que os intervalos de "posse_de_veiculo" são mais amplos em ambas as categorias (falsas e verdadeiras) em comparação com os intervalos de "posse_de_imovel". Isso sugere que as variações nas respostas estão mais dispersas em "posse_de_veiculo", o que pode ser influenciado pela diferença nas médias e/ou pela quantidade de dados.

Portanto, com base nos intervalos de confiança mais amplos para "posse_de_veiculo", a variável "posse_de_veiculo" pode ter uma relevância mais marcante, considerando a variabilidade dos dados. No entanto, é importante lembrar que a análise de diferença estatisticamente significativa também depende de outros fatores, como tamanho da amostra, distribuição dos dados e métodos estatísticos aplicados.
