### 2. Explique com suas palavras o Random forest

O processo do Random Forest segue um passo a passo que envolve bootstrapping, seleção de características, modelagem com árvores de decisão e agregação das decisões. Primeiro, é realizado o bootstrapping, que consiste na criação de múltiplos conjuntos de dados de treinamento a partir do conjunto original, cada um com diferentes subconjuntos de recursos disponíveis. Em seguida, ocorre a seleção de características, onde são escolhidas as características mais relevantes para o treinamento das árvores de decisão.

Após a seleção das características, as árvores de decisão são modeladas utilizando os conjuntos de dados de treinamento gerados pelo bootstrapping. Cada árvore de decisão captura diferentes perspectivas e padrões nos dados, contribuindo para a diversidade do modelo.

Finalmente, ocorre a fase de agregação, onde as decisões de todas as árvores individuais são combinadas para produzir uma previsão final. Essa agregação reduz a variabilidade e melhora a generalização do modelo, tornando o Random Forest uma técnica eficaz para problemas de aprendizado de máquina.

### 3. Qual a diferença entre Bagging e Random Forest?

Bagging e Random Forest são ambos métodos de ensemble que utilizam várias árvores de decisão, mas possuem algumas diferenças significativas:

* Bootstrap:

_Bagging (Bootstrap Aggregating)_: No Bagging, cada árvore de decisão é treinada em um conjunto de dados de treinamento gerado por bootstrapping, onde várias amostras são criadas a partir do conjunto de dados original com reposição.

_Random Forest_: No Random Forest, além de usar bootstrapping, há também uma seleção aleatória de características em cada divisão da árvore de decisão. Isso adiciona uma camada extra de aleatoriedade ao processo, tornando as árvores mais diversificadas.

* Seleção de Características:

_Bagging_: A seleção de características é geralmente feita considerando todas as características disponíveis em cada divisão da árvore.

_Random Forest_: Em cada divisão da árvore, uma amostra aleatória de características é considerada para decidir a melhor divisão. Isso ajuda a reduzir a correlação entre as árvores e a aumentar a diversidade do modelo.

* Número de Árvores:

_Bagging_: Pode ser usado com qualquer número de árvores, geralmente criando um grande número delas para reduzir a variância do modelo.

_Random Forest_: É um tipo específico de Bagging que utiliza um número fixo de árvores, geralmente mais do que o necessário para um Bagging tradicional, devido à seleção aleatória de características em cada árvore.

* Desempenho:

_Bagging_: Geralmente tem um desempenho robusto em uma ampla variedade de conjuntos de dados, reduzindo o overfitting e melhorando a generalização.

_Random Forest_: Por causa da seleção aleatória de características, o Random Forest tende a ter um desempenho ainda melhor que o Bagging em muitos casos, especialmente em conjuntos de dados com muitas características.
