### 2. Explique com suas palavras o Random forest

O processo do Random Forest segue um passo a passo que envolve bootstrapping, seleo de caractersticas, modelagem com rvores de deciso e agregao das decises. Primeiro,  realizado o bootstrapping, que consiste na criao de mltiplos conjuntos de dados de treinamento a partir do conjunto original, cada um com diferentes subconjuntos de recursos disponveis. Em seguida, ocorre a seleo de caractersticas, onde so escolhidas as caractersticas mais relevantes para o treinamento das rvores de deciso.

Aps a seleo das caractersticas, as rvores de deciso so modeladas utilizando os conjuntos de dados de treinamento gerados pelo bootstrapping. Cada rvore de deciso captura diferentes perspectivas e padres nos dados, contribuindo para a diversidade do modelo.

Finalmente, ocorre a fase de agregao, onde as decises de todas as rvores individuais so combinadas para produzir uma previso final. Essa agregao reduz a variabilidade e melhora a generalizao do modelo, tornando o Random Forest uma tcnica eficaz para problemas de aprendizado de mquina.

### 3. Qual a diferena entre Bagging e Random Forest?

Bagging e Random Forest so ambos mtodos de ensemble que utilizam vrias rvores de deciso, mas possuem algumas diferenas significativas:

* Bootstrap:

_Bagging (Bootstrap Aggregating)_: No Bagging, cada rvore de deciso  treinada em um conjunto de dados de treinamento gerado por bootstrapping, onde vrias amostras so criadas a partir do conjunto de dados original com reposio.

_Random Forest_: No Random Forest, alm de usar bootstrapping, h tambm uma seleo aleatria de caractersticas em cada diviso da rvore de deciso. Isso adiciona uma camada extra de aleatoriedade ao processo, tornando as rvores mais diversificadas.

* Seleo de Caractersticas:

_Bagging_: A seleo de caractersticas  geralmente feita considerando todas as caractersticas disponveis em cada diviso da rvore.

_Random Forest_: Em cada diviso da rvore, uma amostra aleatria de caractersticas  considerada para decidir a melhor diviso. Isso ajuda a reduzir a correlao entre as rvores e a aumentar a diversidade do modelo.

* Nmero de rvores:

_Bagging_: Pode ser usado com qualquer nmero de rvores, geralmente criando um grande nmero delas para reduzir a varincia do modelo.

_Random Forest_:  um tipo especfico de Bagging que utiliza um nmero fixo de rvores, geralmente mais do que o necessrio para um Bagging tradicional, devido  seleo aleatria de caractersticas em cada rvore.

* Desempenho:

_Bagging_: Geralmente tem um desempenho robusto em uma ampla variedade de conjuntos de dados, reduzindo o overfitting e melhorando a generalizao.

_Random Forest_: Por causa da seleo aleatria de caractersticas, o Random Forest tende a ter um desempenho ainda melhor que o Bagging em muitos casos, especialmente em conjuntos de dados com muitas caractersticas.
